# DATA ANALYSIS ON TITANIC DATASET

```python
import pandas as pd
import numpy as np
```

```python
# loading the dataset

data = pd.read_csv('titanic.csv')
```

```python
# first five rows

data.head()
```

## Checking and Handling Missing Values

```python
# checking if there are any missing(null) values

data.isnull().sum()
```

### Handling missing values

```python
data['Age'] = data['Age'].fillna(0)

data

# sets the Null age to 0
```

## Dropping Column

```python
# dropping the rows in which ' Cabin ' column has null values

data = data.dropna()
```

```python
# again checking for missing values after cleaning

data.isnull().sum()
```

# Calculating Basic Statistics for Numerical Variables (Age & Fare)

```python
#

stat_values = data.describe()

stat_values
```

```python
# calculating median for only "numeric columns"

median = data.median(numeric_only=True)

stat_values.loc['median'] = median


stat_values
```

## DATA VIZUALIZATION USING MATPLOTLIB

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Histogram for Age
# vizualizes frequncy of age-group people

plt.figure(figsize=(6, 4))
plt.hist(data['Age'], bins=30, edgecolor='black')  
plt.title('Histogram for Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```
